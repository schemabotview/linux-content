## What is Linux?

If you ask ten people what Linux is, you'll get ten slightly different answers — and confusingly, all of them are correct. Let's untangle that.

**Linux**, strictly speaking, is just one program: the **kernel**. The kernel is the piece of software that talks directly to your computer's hardware — the CPU, memory, disks, network cards. Every time a program on your machine wants to read a file or print to the screen, it ultimately asks the kernel to do it.

But you can't *use* just a kernel. A kernel by itself has no command prompt, no programs to run, no way for you to interact with it. You need a bunch of other software wrapped around it — a shell to type into, basic commands like `ls` and `cp`, a way to install more software, and so on.

When all that is bundled together, you get a **Linux distribution** (or "distro" for short). When people say "I run Linux," they almost always mean "I run a Linux distribution" — Ubuntu, Debian, Fedora, RHEL, Arch, and so on. Each distribution makes slightly different choices about which programs to include, how to organise things, and how to install new software, but underneath they all run the same Linux kernel.

A useful mental picture:

```
  Your programs (firefox, python, your scripts)
  ──────────────────────────────────────────────
  The userspace tools (ls, cp, bash, systemctl)    ← the distribution adds these
  ──────────────────────────────────────────────
  Linux — the kernel                                ← what the word "Linux" technically means
  ──────────────────────────────────────────────
  Your hardware (CPU, RAM, disk, network)
```

For this course, you can use just about any major distribution. We'll show examples on **Ubuntu** (the most common starting point) and mention where Red Hat-family distros (RHEL, Fedora, Rocky, Alma) differ.

## Why Linux matters

You may already use Linux without realising it. Here's where it lives:

- **Almost every server on the internet runs Linux.** Web servers, databases, mail servers, content delivery networks — the overwhelming majority are Linux. Amazon, Google, Facebook, Netflix, GitHub: Linux underneath.
- **Every Android phone is built on Linux.** Android's kernel is Linux with some additions.
- **The cloud is Linux.** AWS, Google Cloud, Azure, DigitalOcean — when you spin up a "virtual machine," it's almost certainly Linux.
- **Containers are Linux.** Docker, Kubernetes, all of it runs on Linux features.
- **Embedded devices.** Routers, smart TVs, car infotainment systems, the Mars rovers — Linux.
- **Supercomputers.** Literally all of the top 500 supercomputers in the world run Linux.

What this means for you: if you work anywhere near software — as a developer, sysadmin, DevOps engineer, data engineer, security professional — knowing Linux isn't optional. It's the floor.

## Getting your own Linux to practise on

You can read about Linux all day, but you only learn it by typing. Here are the four common ways to get a Linux prompt to play with, ordered from easiest to most involved.

**Option 1 — Cloud VM (recommended).** Spin up a small Linux server on a cloud provider. DigitalOcean, AWS Lightsail, Linode, Hetzner — all charge a few dollars a month for a tiny "droplet" or VM that's perfect for learning. You SSH in (covered in notebook 08) and you have a real Linux system.

- Pros: it's a *real* Linux server, you can break it and rebuild it, you'll learn SSH naturally.
- Cons: needs an internet connection, costs a few dollars.

**Option 2 — Multipass (free, runs Ubuntu locally).** Canonical's [Multipass](https://multipass.run) gives you a one-command Ubuntu VM on your laptop. Free, no signup, works on macOS, Windows, and Linux hosts.

```bash
multipass launch --name linux-lab
multipass shell linux-lab
```

- Pros: free, fast to start, isolated VM (can't break your host).
- Cons: only Ubuntu, uses a few GB of RAM.

**Option 3 — WSL2 (Windows users).** If you're on Windows 10/11, Windows Subsystem for Linux 2 gives you a real Linux kernel running alongside Windows. Open Microsoft Store, install "Ubuntu", launch.

- Pros: zero friction on Windows, files are shared between Linux and Windows.
- Cons: a few systemd-related quirks (mostly fixed in WSL2's recent versions).

**Option 4 — Docker container.** If you already have Docker, the fastest "I just want a Linux prompt right now" is:

```bash
docker run -it --rm ubuntu:24.04 bash
```

- Pros: instant, disposable, no setup.
- Cons: missing systemd by default (limits notebook 09), no persistence unless you mount a volume.

**Option 5 — VirtualBox or VMware.** Download an Ubuntu ISO, install in a desktop VM. Heavyweight compared to the above but gives you a full graphical desktop.

**For this course**, the path of least friction is **Multipass** (if you're on a laptop) or a **cheap cloud VM** (if you want the most authentic experience). Pick one and stick with it — switching environments mid-course is a distraction.

## What you see when you log in

The first time you log into a Linux system, you'll see a **prompt** — a line that ends in `$` or `#` and is waiting for you to type. It looks something like this:

```
ubuntu@linux-lab:~$
```

Read that left to right:

- **`ubuntu`** — your username (the account you logged in as).
- **`@`** — separator, literally an "at" sign.
- **`linux-lab`** — the hostname (the name of this machine).
- **`:`** — separator.
- **`~`** — the current directory. The tilde is shorthand for "your home directory" (we'll get to home directories in a moment).
- **`$`** — the prompt symbol. **`$`** means you're a regular user; **`#`** means you're the **root** user (the administrator), and you need to be careful.

So `ubuntu@linux-lab:~$` reads as: "user `ubuntu`, on machine `linux-lab`, currently in the home directory, ready for your command."

The prompt is generated by the **shell** — the program that's actually reading what you type and running it. On most Linux systems the shell is called **bash**. Other shells exist (zsh, fish, dash), but bash is the default and what we'll use throughout.

When you type a command and press Enter, the shell:

1. Reads your line.
2. Figures out what command you meant.
3. Runs that command.
4. Prints the output (if any).
5. Shows you a fresh prompt and waits again.

That loop — **read, run, print, prompt again** — is your conversation with Linux. Everything in this course is variations on that loop.

## Listing what's in a directory — `ls`

A **directory** in Linux is the same idea as a **folder** in Windows or macOS — a container that holds files and other directories. The command to list what's in one is **`ls`** (short for "list").

By itself, `ls` lists the current directory. Give it a path and it lists that path:

- `ls` — list the current directory.
- `ls /etc` — list the `/etc` directory.
- `ls /` — list the top-level (root) of the filesystem.

`ls` accepts a handful of flags that change how it shows things. The ones you'll use constantly:

- **`-l`** ("long format") — show one file per line, with permissions, owner, size, modification time.
- **`-a`** ("all") — also show hidden files (those starting with a dot, like `.bashrc`).
- **`-h`** ("human") — show file sizes in K/M/G instead of raw bytes. Only useful combined with `-l`.
- **`-t`** — sort by modification time, newest first.
- **`-r`** — reverse the sort.

Flags can be combined: `ls -lah` is the same as `ls -l -a -h`. You'll see `ls -la` and `ls -lh` constantly — they're worth memorising.

````python
!ls
!ls -la /etc | head -10
!ls -lh /var/log 2>/dev/null | head -10
````

## Moving around — `cd`, `..`, `.`, `~`

To change which directory you're "in," use **`cd`** ("change directory"). A few magic shorthands save constant typing:

| Shortcut | Means |
|---|---|
| `~` | your home directory (e.g. `/home/ubuntu`) |
| `.` | the current directory |
| `..` | the parent directory (one level up) |
| `-` | the previous directory you were in (toggles back and forth) |
| `/` | the root of the filesystem |

Examples:

- `cd /etc` — go to `/etc`.
- `cd ..` — go up one level (e.g. from `/etc/apache2` to `/etc`).
- `cd ~` or just `cd` — go to your home directory.
- `cd -` — toggle back to where you just were.
- `cd /` — go to the top.

After every `cd`, use `pwd` to confirm where you are if you're not sure.

**Absolute vs relative paths.** A path starting with `/` is **absolute** — it specifies the location from the root of the filesystem. A path that doesn't start with `/` is **relative** — it's interpreted from wherever you currently are. So if you're in `/var`, then `cd log` (relative) takes you to `/var/log`, and `cd /log` (absolute) would try to take you to a top-level `/log` (which probably doesn't exist).

````python
!pwd
!cd /tmp && pwd
!cd /etc && pwd
!cd .. && pwd
!cd ~ && pwd
````

## Command structure — the anatomy of a command

Every Linux command follows the same shape:

```
command   [options]   [arguments]
```

- **`command`** — the program to run (`ls`, `cp`, `grep`, etc.).
- **`options`** (also called **flags** or **switches**) — modify how the command behaves. Start with `-` (short form, one letter, like `-l`) or `--` (long form, full word, like `--long`).
- **`arguments`** — what the command operates on. Usually filenames, paths, or text.

Examples, broken down:

```
ls -l /etc/passwd
│  │  └── argument: the file to list
│  └── option: -l (long format)
└── command: ls

cp -r /home/ubuntu/work /backup
│  │  └────── two arguments: source and destination
│  └── option: -r (recursive)
└── command: cp
```

Short and long forms often coexist. `ls -a` and `ls --all` do the same thing. Some options take their own values:

```
grep --max-count=5 'error' /var/log/syslog
│    └─────────── option with a value (--max-count=5)
│                                       │       │
│                                       │       └── argument: file to search
│                                       └── argument: pattern to search for
└── command: grep
```

Order tip: most commands accept options anywhere in the line, but **the convention** is `command options arguments`. Stick to that for readability.

A `--` on its own marks the end of options. After it, everything is treated as an argument, even if it starts with `-`. That's how you grep for a literal dash: `grep -- '-x' file`.

````python
!ls -l --human-readable /etc/passwd /etc/hostname
````

## A first tour of the filesystem hierarchy

Unlike Windows, Linux doesn't have drive letters (`C:`, `D:`). Everything — every disk, every device, every file — lives under a single tree rooted at **`/`** ("the root").

A handful of top-level directories you'll meet over and over:

| Path | What it's for |
|---|---|
| **`/home`** | user home directories. Your stuff lives in `/home/<your-username>/`. |
| **`/root`** | the root (administrator) user's home — deliberately not under `/home`. |
| **`/etc`** | system configuration files. Plain text, editable. "All my settings live here." |
| **`/var`** | files that change while the system runs — logs (`/var/log`), mail spool, package caches. |
| **`/tmp`** | temporary files. Anyone can write here. Usually wiped on reboot. |
| **`/usr`** | most installed software lives here — `/usr/bin` for binaries, `/usr/lib` for libraries. |
| **`/bin`, `/sbin`** | essential commands and admin commands. On modern distros, often symlinks into `/usr/bin` and `/usr/sbin`. |
| **`/opt`** | software installed outside the package manager — third-party apps. |
| **`/boot`** | the Linux kernel image and bootloader config. |
| **`/dev`** | special files representing devices (`/dev/sda` = first disk, `/dev/null` = the bit bucket). |
| **`/proc`** | a *virtual* directory that exposes running-process information. Not on disk. (More in notebook 05.) |
| **`/sys`** | another virtual directory exposing kernel and device info. |
| **`/mnt`, `/media`** | mount points for additional filesystems (USB drives, network shares). |

The shorthand `~` always means **your** home directory. So if your username is `ubuntu`, `~` is the same as `/home/ubuntu`. If you `cd ~/Documents` you're going to `/home/ubuntu/Documents`.

Don't try to memorise everything. Just remember: **config goes in `/etc`, your stuff goes in `/home/<you>`, logs go in `/var/log`, temp scratch goes in `/tmp`.** The rest you'll meet as you need them.

````python
!ls /
!ls /etc | head -10
!ls /var
````

## Tab completion and history — the two shortcuts you'll use most

Two features of the shell that will save you more keystrokes than anything else.

**Tab completion.** Start typing a command or a path and press **Tab**. The shell will complete it for you if there's only one possibility. Press **Tab twice** if there are multiple options — it'll show you the list.

Try it: type `ls /et` and press Tab. The shell completes to `ls /etc/`. Type `cd /usr/sh` and press Tab — completes to `/usr/share/`. Press Tab twice after `cd /` and you'll see all top-level directories.

Tab completes commands, paths, filenames, and even some option names. Use it constantly. Typing `/usr/share/applications/firefox.desktop` by hand is a mistake; tab-completing through it is fast and typo-proof.

**Command history.** Every command you run is recorded. Recall previous commands with the up arrow (↑) and down arrow (↓). To jump back faster:

- **`history`** — show the last N commands (with line numbers).
- **`!!`** — re-run the previous command. Useful as `sudo !!` when you forgot `sudo`.
- **`!42`** — re-run command number 42 from `history`.
- **`!ls`** — re-run the most recent command starting with `ls`.
- **Ctrl-R** — interactive reverse-search. Press it, start typing, and it finds the most recent matching command. Press Ctrl-R again to step backwards through matches. Hit Enter to run, or Tab/right-arrow to edit.

History persists across sessions in `~/.bash_history`. By default, bash keeps the last 1000 commands.

````python
!history | tail -10
````

## Getting help — `man`, `--help`, `apropos`, `info`

You don't need to memorise every command. You need to know how to **look up** what you've forgotten. Linux ships with three layers of help, all available locally without an internet connection.

**`man <command>`** — the **manual page** for a command. The most thorough reference. Type `man ls` and you get pages of options, examples, and notes. While inside `man`:

- **Space** or **PgDn** — next page.
- **`b`** or **PgUp** — previous page.
- **`/text`** — search forwards for `text`.
- **`n`** — next search match. **`N`** — previous match.
- **`q`** — quit back to the shell.

Manual pages are organised into numbered sections:

| Section | What's there |
|---|---|
| 1 | User commands (`man 1 ls`, `man ls`) |
| 5 | File formats (`man 5 passwd` — the format of `/etc/passwd`) |
| 8 | System administration commands (`man 8 mount`) |

When a name lives in multiple sections, you can specify: `man 5 passwd` is the file format; `man 1 passwd` is the password-changing command. They are *different topics*.

**`<command> --help`** — almost every command supports `--help` (or sometimes `-h`) and prints a short summary built into the program itself. Faster than `man` for a quick reminder.

**`apropos <keyword>`** — search the descriptions of all installed man pages by keyword. The "I don't know which command, I just know what I want" tool. `apropos compress` lists every command involving compression.

**`info <command>`** — a more verbose alternative to `man`, used by GNU tools. `info coreutils` is genuinely excellent for the basic commands.

The flow when you're stuck: try `--help` first (quickest), then `man` (most thorough), then `apropos` if you don't know the command name. After a while you'll have most common commands memorised, but the docs are always there.

````python
!ls --help | head -15
!man -f passwd 2>/dev/null
!apropos --search-syntax 2>/dev/null | head -3 || apropos compress 2>/dev/null | head -5
````
